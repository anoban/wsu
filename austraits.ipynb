```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
austraits = pd.read_csv(r"./austraits-3.0.2/traits.csv", low_memory=False)
```

```python
austraits.loc[:, "trait_name"][austraits.loc[:, "trait_name"].str.contains("pathway")]
```

```python
multi_pathway_species = austraits.query("trait_name == 'photosynthetic_pathway'").loc[:, ["taxon_name", "value"]].dropna().\
           sort_values(by="taxon_name").drop_duplicates().groupby("taxon_name").count().query("value > 1").index
multi_pathway_species
```

```python
with pd.option_context("display.max_rows", None):
    print(
        austraits.loc[austraits.taxon_name.isin(multi_pathway_species), ["taxon_name", "trait_name",
              "value"]].query("trait_name == 'photosynthetic_pathway'").sort_values(by="taxon_name").drop_duplicates().loc[:,
        ["taxon_name", "value"]]
    )
```
