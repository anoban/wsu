```python
fig, axes = plt.subplots(nrows=1, ncols=16)
fig.set_size_inches(32, 2)
x = y = 0
for (i, axis) in enumerate(axes.flatten()):
    axis.set_xticks([])
    axis.set_yticks([])
    axis.imshow(roots[x:x+178, y:y+178])
    axis.set_xlabel(f"{i}")
    x += 178
    if not (x % 712):
        y += 178
        x = 0
    
plt.tight_layout()
# plt.show()
plt.savefig(fname=r"./root_tokens.jpeg", format="jpeg", dpi=400)
```
