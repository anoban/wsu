## ___Genera with more than one photosynthetic pathway___
--------------------------

```python
australian_photosynthetic_pathways = austraits.query("trait_name == 'photosynthetic_pathway'").\
                                            loc[:, ["taxon_name", "value"]].dropna()
australian_photosynthetic_pathways
```

```python
australian_photosynthetic_pathways.loc[:, "genus"] = \
            australian_photosynthetic_pathways.taxon_name.apply(lambda _: _.split(' ')[0].strip())
```

```python
for (genus, df) in australian_photosynthetic_pathways.groupby("genus"):
    if df.value.unique().size > 1:
        print(genus, np.unique(df.value, return_counts=True))
```
